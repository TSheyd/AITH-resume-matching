```python
import fasttext
import fasttext.util
```

```python
def embeddings_ft(data, shape, feature_name):
    model = fasttext.load_model('../../rabota_ru/cc.ru.300.bin')
    fasttext.util.reduce_model(model, shape)
    
    emb_vector = []
    
    data = data.tolist()
    
    for words in data:
        vec = [model.get_word_vector(word) for word in words.split(' ')][0].tolist()
        emb_vector.append(vec)
    return pd.DataFrame(emb_vector, columns=[feature_name+'_{}'.format(feature) for feature in range(1, shape+1)])
```

```python
import pandas as pd
import numpy as np
import fasttext
import fasttext.util
from sklearn.neighbors import NearestNeighbors
from joblib import dump, load
```

```python
model = fasttext.load_model('model/cc.ru.300.bin')
fasttext.util.reduce_model(model, 100)
```

```python
df = pd.read_csv('data/hhparser_vacancy_short.csv')
```

```python
df
```

```python
def embeddings_ft(data, model, feature_name):
    emb_vector = []
    
    data = data.tolist()
    
    for words in data:
        vec = [model.get_word_vector(word) for word in words.split(' ')][0].tolist()
        emb_vector.append(vec)
    return pd.DataFrame(emb_vector, columns=[feature_name+'_{}'.format(feature) for feature in range(1, 100+1)])
```

```python
emb_vac = embeddings_ft(df['name'], model, 'vac')
```

```python
emb_area = embeddings_ft(df['area_name'], model, 'city')
```

```python
df_emb = pd.concat([df.reset_index(drop=True), 
                            emb_vac, 
                            emb_area], axis=1)
```

```python
df_emb['salary_from'] = df_emb['salary_from'].fillna(0)
```

```python
df_emb.to_csv('data/edited.csv', index=False)
```

```python
df_emb[['vac_' + str(i+1) for i in range(100)]+['city_' + str(i+1) for i in range(100)]+['salary_from']]
```

```python
knn = NearestNeighbors(n_neighbors=150).fit(df_emb[['vac_' + str(i+1) for i in range(100)]+['city_' + str(i+1) for i in range(100)]+['salary_from']])
```

```python
dump(knn, 'model/knn.joblib')
```
